```python
import numpy as np
import random
import pylab

#난수를 저장할 배열
randomArr=[]

# 1. 0~9사이의 난수를 100개 생성해서 배열에 저장
for i in range(100):
    randomArr.append(np.random.randint(0,10))

print "random array : %s" %randomArr

# 2. 평균 구하기
arrMean = np.mean(randomArr)

# 3. 표준편차 구하기
arrSD = np.std(randomArr)

# 4. 분산 구하기
arrVar = np.var(randomArr)

print "mean : %s" %arrMean
print "standard deviation : %s" %arrSD
print "variance : %s" %arrVar

# 5. 시계열 그래프 그리기
pylab.figure(1)
pylab.plot(range(100),randomArr)
pylab.title('(A) time series')
pylab.xlabel('index')
pylab.ylabel('random number')

# 6. 히스토그램 그리기
pylab.figure(2)
pylab.title('(B) histogram')
pylab.xlabel('random number')
pylab.ylabel('frequency')
pylab.hist(randomArr)


pylab.show()

print('(A) 데이터의 패턴이 불규칙하므로 데이터의 랜덤성을 알 수 있음')
print('(B) 0~9사이의 정수가 몇 번 생성되었는지를 나타냄')
```
